# Star Predection

The dataset yelp_business_official_test_empty.csv contains 8 new businesses that are not
found in dataset provided by Yelp. These 8 businesses are missing their star rating. Using what
we’ve learned in this course, you are to “predict” the stars for these 8 new businesses.

For example, if a new business was a fast food restaurant in Austin, TX, you might “predict” the
star rating by assigning the average rating for similar restaurants in Austin, TX.

```python
import pandas as pd
import numpy as np
```

```python
predict_stars = pd.read_csv('data-sets/yelp_business_official_test_empty.csv')
```

```python
#See what we're dealing with
predict_stars
```

```python
#dropping columns that consist of all NAs in predict_stars and corresponding ones in business. Business_id column is being dropped too
predict_stars.dropna(axis=1, how='all', inplace=True)
predict_stars.drop(columns='business_id', inplace=True)
```

```python
#Open data file. Use columns from business that are included in given predict_stars dataframe.
#Open data file again but this time just the ratings and index columns
business = pd.read_csv('data-sets/yelp_academic_dataset_business.csv', usecols=predict_stars.columns)[predict_stars.columns]
ratings = pd.read_csv('data-sets/yelp_academic_dataset_business.csv', usecols=['stars'])
```

```python
business.info()
```

```python
ratings.info()
```

```python
#clean predict_stars
predict_stars.name = predict_stars.name.str.replace('"', '')
predict_stars.address = predict_stars.address.str.replace('"', '')
predict_stars.categories = predict_stars.categories.str.replace(';', ', ')
predict_stars
```

```python
#Sorting the columns so that the 'more important' rows are on the left. This will make more sense later (hopefully)

cols=predict_stars.columns
newCols = list(cols[[0, 9, 1, 4, 2, 3, 7, 8, 5, 6]])
newCols
predict_stars = predict_stars.reindex(columns=newCols)
business = business.reindex(columns=newCols)
predict_stars
```

```python
#For each cell of predict_stars, finds a match in business dataframe. 
#Prints each time matches are found and adds it to a temporary dataframe. 
#At the end of each row, this dataframe is copied into 'resultList'
#Also creates another list of lists. The outer list represents each row, and the inner represents each matched columns for that row



rowFoundList = [] #Outer index list
resultList = [] #Dataframe list

for i in range(len(predict_stars)):
    foundList = [] #Inner index list
    compareDF = pd.DataFrame(columns=predict_stars.columns) #Sets columns of temp dataframe
    
    for j in range(len(predict_stars.columns)):
        if(j == 7):#Skip is_open column. Uneeded matches
            continue
            
        compareDF = pd.concat([compareDF, business.loc[business.iloc[:,j] == predict_stars.iloc[i, j]]]) #Adds matches to temp dataframe as new rows
        if(len(business.loc[business.iloc[:,j] == predict_stars.iloc[i, j]]) > 0): #If(match was found)
            print("Row {}: Found match(es) in {}".format(i , predict_stars.columns[j])) #Print statement saying match found
            foundList.append(j) #Add column index of match to inner list
    compareDF = compareDF.merge(ratings, how='left', left_index=True, right_index=True) #Merge matches with rating numbers from ratings dataframe
    compareDF.drop_duplicates(inplace=True) #Drop any duplicates
    resultList.append( compareDF.copy()) #Add column index to list 
    print("\n-----------------------------------\n")
    rowFoundList.append(foundList) #Add list of column indexes to outer row index list
```

```python
#To show results, printing off Row 0 and Row 4 resulting matches
resultList[0]
```

```python
resultList[4]
```

We see that results row 0 just has matches in the categories column, while results row 4 has matches in the name, review_count, and categories columns. Now we need to filter these down so only the most similar data is left. In row 4's case this would mean finding if there are any rows that match all 4 of matched columns.

```python
#This is probably pretty convoluted but here goes nothing...
#Takes the resultList as input from earlier. That list contains all matches in all columns for each row of the original predict_stars DF
#ResultList is structured as 2D list so outer list corresponds to each row, while inner list contains each row of business that matched in any column with that row of predict_stars
#Output will be saved list. This will again be 2D list. Outer list corresponds to each row of predict_stars
    #Inner list contains filtered matches for that row
    
#Filtered matches meaning that it starts on left most columns matches and stores the boolean array in truth. 
#It then goes one column to the right (which is why I sorted earlier) and compares that with the previous boolean array.
#If any are left 'True' then that means there is a match in both categories and it is saved in 'saved'
#If there are no matches, then the previous truth array is used again, retrieved from 'lastTruth'

#If there is only matched from a single column, then at the end it shall see that 'saved' is empty and just use the original input for that row as the row's output

savedList = []
saved = pd.DataFrame(columns=predict_stars.columns)

for outerRow in range(len(resultList)):
    tempRowDF = resultList[outerRow]
    truth = True
    for innerRow in rowFoundList[outerRow]:
        lastTruth = truth
        truth = (tempRowDF.iloc[:,innerRow] == predict_stars.iloc[outerRow, innerRow]) & truth
        if(len(tempRowDF.loc[truth]) > 0):
            saved = tempRowDF.loc[truth]
            continue
        else:
            truth=lastTruth
    if(len(saved) > 0):
        savedList.append(saved)
    else:
        savedList.append(tempRowDF)
```

```python
#Finally, just iterate through savedList to get remaining data rows and then calculate the average star rating
#Outputing them in starList. Then reading in the original unedited predict_stars DF and setting the star column

starList = []

for i in range(len(savedList)):
    starList.append(savedList[i].stars.mean())

    
predict_stars_original = pd.read_csv('data-sets/yelp_business_official_test_empty.csv')
predict_stars_original.stars = starList
predict_stars_original.stars
```
